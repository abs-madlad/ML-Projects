```python
td['mainroad'] = td['mainroad'].replace({'yes': 1, 'no': 0})
td['guestroom'] = td['guestroom'].replace({'yes': 1, 'no': 0})
td['basement'] = td['basement'].replace({'yes': 1, 'no': 0})
td['hotwaterheating'] = td['hotwaterheating'].replace({'yes': 1, 'no': 0})
td['airconditioning'] = td['airconditioning'].replace({'yes': 1, 'no': 0})
td['prefarea'] = td['prefarea'].replace({'yes': 1, 'no': 0})
td['furnishingstatus'] = td['furnishingstatus'].replace({'furnished': 2, 'semi-furnished': 1, 'unfurnished':0})
```

```python
correlation_matrix= td.corr()
plt.figure(figsize=(15,10))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt='.2f')
plt.title('Correlation matrix heatmap')
plt.show()
```

```python
x= td.drop(['price'], axis=1)
y= td['price']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2, random_state= 123)
```

```python
r2_list= []
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

model= LinearRegression()
model.fit(x_train,y_train)

#predictions= model.predict(x_test_scale)
ytrain_pred = model.predict(x_test)
mse = mean_squared_error(y_test, ytrain_pred)
mae = mean_absolute_error(y_test, ytrain_pred)
r2 = r2_score(y_test, ytrain_pred)
print('MSE',np.sqrt(mse))
print('MAE',mae)
print('R2Score',r2)
r2_list.append(['Linear Regression',r2])
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr= RandomForestRegressor(n_estimators=200, random_state=40)

rfr.fit(x_train, y_train)

predictions = rfr.predict(x_test)

mse = mean_squared_error(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
print('MSE',np.sqrt(mse))
print('MAE',mae)
print('R2Score',r2)
r2_list.append(['Random Forest Regression',r2])
```

```python
from sklearn.cross_decomposition import PLSRegression
pls_model = PLSRegression(n_components=2).fit(x_train, y_train)
```

```python
predictions= pls_model.predict(x_test)
mse = mean_squared_error(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)
print('MSE',np.sqrt(mse))
print('MAE',mae)
r2= r2_score(y_test, predictions)
print('R2_score',r2)
r2_list.append(['PLS Regression',r2])
```

```python
from sklearn.linear_model import Lasso
las= Lasso(alpha=0.5).fit(x_train,y_train)
predictions=las.predict(x_test)

mse = mean_squared_error(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)

print('MSE',np.sqrt(mse))
print('MAE',mae)
r2= r2_score(y_test, predictions)
print('R2_score',r2)
r2_list.append(['Lasso Regression',r2])
```

```python
print('R2_Score')
for i in r2_list:
  print(i[0],'--->',i[1])
```
